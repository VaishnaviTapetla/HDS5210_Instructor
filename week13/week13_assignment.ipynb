### 1. Basic Statistics

Use Python to read in your data file and show many rows and columns your data has.

### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
unique_counts = data.nunique()

# Sort the counts in descending order
sorted_counts = unique_counts.sort_values(ascending=False)

# Print unique value counts for all columns
print("Number of unique values in each column:")
print("-" * 40)
for col, count in sorted_counts.items():
    print(f"{col}: {count} unique values")

# Find column with most unique values
most_diverse_col = sorted_counts.index[0]
most_diverse_count = sorted_counts.iloc[0]
second_diverse_col = sorted_counts.index[1]
second_diverse_count = sorted_counts.iloc[1]

print("\nMost diverse column:")
print(f"{most_diverse_col} with {most_diverse_count} unique values")

print("\nSecond most diverse column:")
print(f"{second_diverse_col} with {second_diverse_count} unique values")
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
import matplotlib.pyplot as plt
numeric_columns = data.select_dtypes(include=['int64', 'float64']).columns

# Print numeric columns so we can see what we're working with
print("Numeric columns in the dataset:")
print(numeric_columns.tolist())

# Create histograms for each numeric column
for column in numeric_columns:
    plt.figure(figsize=(10, 6))
    plt.hist(data[column], bins=30)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.show()

    # Print basic stats for each column
    print(f"\nStatistics for {column}:")
    print(f"Mean: {data[column].mean():.2f}")
    print(f"Median: {data[column].median():.2f}")
    print(f"Min: {data[column].min():.2f}")
    print(f"Max: {data[column].max():.2f}")
    print("-" * 40)
```
